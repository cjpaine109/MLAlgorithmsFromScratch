## Building a - Polynomial Regression - model from scratch

```python
# IMPORTS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the data
data = pd.read_csv("Dataset/Position_Salaries.csv")
data.head()
```

```python
# Plot the data points
plt.scatter(data.Level, data.Salary, color='blue', label='Data Points')
plt.plot(data.Level, data.Salary, 'r--', label="Ideal Line of Best Fit")
plt.title("Salary Dataset")
plt.legend()
plt.show()
```

```python
# Lin-reg
X = data[['Level']].values
y = data[['Salary']].values
X_b = np.hstack([np.ones((X.shape[0], 1)), X])

def compute_params(X, y):
    theta = np.linalg.inv(X.T @ X) @ (X.T @ y)
    return theta

def predict(X, theta):
    return X @ theta

# Call these functions and plot results
theta = compute_params(X_b, y)
preds = predict(X_b, theta)

plt.scatter(data.Level, data.Salary, color='blue', label='Data Points')
plt.plot(data.Level, preds, 'r--', label="Line of Best Fit")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
# Poly-reg
degree =  3
X_poly = np.hstack([X**i for i in range(1, degree + 1)])
X_poly = np.hstack([np.ones((X.shape[0], 1)), X_poly])

# Call the functions and test the model
theta = compute_params(X_poly, y)
print(theta)
poly_preds = predict(X_poly, theta)

plt.scatter(data.Level, data.Salary, color='Blue', label='Data Points')
plt.plot(data.Level, poly_preds, 'r--', label="Line of Best Fit")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```
