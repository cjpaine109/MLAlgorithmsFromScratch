## Linear Regression ---> from scratch

```python
# Building a simple-linear regression model to fit some randomly generated data points -> closed-form solution ('The Normal Equation')
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) # column vector
y = 4 + 3 * X + np.random.randn(m, 1) # column vector | y = 4 + 3x + noise

# Visualize this data
plt.scatter(X, y)
plt.show()
```

```python
# Transform the input data
X = np.hstack([np.ones((X.shape[0], 1)), X])
y = np.ravel(np.asanyarray(y))

# Define a fit-function
def fit(X, y):
    compute_params = np.linalg.inv(X.T @ X) @ (X.T @ y) # 'The Normal Equation'
    bias = compute_params[0]
    weights = compute_params[1]
    return bias, weights

bias, weights = fit(X, y)

# Define a predict function
def predict(X):
    bias_weights = np.vstack(([bias], weights)).reshape(-1, 1)
    return X @ bias_weights

# Make predictions
preds = predict(X)
```

```python
# Generate some data
np.random.seed(42)
m = 100
X = 4 * np.random.rand(m, 1) - 2 # column vector
y = 4 + 3 * X + X**2 * 5 + np.random.randn(m, 1) # column vector | y = 4 + 3x + X^2 * 5 + noise (quadratic)

# Visualize this data
plt.scatter(X, y)
plt.show()

# Transform the input data
X = np.hstack([np.ones((X.shape[0], 1)), X])
y = np.ravel(np.asanyarray(y))

# call the functions
bias, weights = fit(X, y)
preds = predict(X)

# Plot the results
np.random.seed(42)
m = 100
X = 4 * np.random.rand(m, 1) - 2 # column vector
y = 4 + 3 * X + X**2 * 5 + np.random.randn(m, 1) # column vector | y = 4 + 3x + X^2 * 5 + noise (quadratic)

# 'New line of best-fit'
plt.scatter(X, y)
plt.plot(X, preds, 'r--')
plt.show()
```
